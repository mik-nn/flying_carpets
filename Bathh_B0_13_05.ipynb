```python
#  Convert 13-05 Label to csv format
import pandas as pd
import os
import cv2
from matplotlib import pyplot as plt
```

```python
SRC = '../Dataset/Batch_BO_13_05-Annotations'
DST = '../Labels_in_Folders'
IMAGE = '../Images'
VIDEO_PATH = '../Video'
COL = ['filename', 'width', 'height', 'class',  'xmin',  'ymin',  'xmax', 'ymax']
#LABELS = ['kover6090', 'kover115200', 'kover150300', 'kover85150', 'kover115400']
LABELS = ['kover6090', 'kover85150', 'kover115200', 'kover150300', 'kover115400']
```

```python
df = pd.DataFrame(columns=COL)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
width = 1920
height = 1080
fps = 25
font = cv2.FONT_HERSHEY_SIMPLEX
out_name = os.path.join(VIDEO_PATH,'Batch_BO_13_05_bb.mp4')
out = cv2.VideoWriter(out_name,fourcc, fps, (width,height))
files = os.listdir(SRC)
files.sort()
print(files)
for file in files:
    in_df = pd.read_csv(os.path.join(SRC,file),sep=' ',names=COL[3:])
    in_df[COL[0]] = file[:-3]+'jpg'
    img_path = os.path.join(IMAGE,file[:-11])+'/'+file[:-3]+'jpg'
    img = cv2.imread(os.path.join(IMAGE,file[:-11],file[:-3]+'jpg'))
    in_df[COL[1]] = img.shape[1]    #width
    in_df[COL[2]] = img.shape[0]    #height
    # DeNormalisation
    x = in_df.xmin
    w = in_df.xmax
    y = in_df.ymin
    h = in_df.xmax
    in_df.xmin = int((x-w/2) * img.shape[1])
    in_df.xmax = int((x+w/2)* img.shape[1])
    in_df.ymin = int((y-h/2) * img.shape[0])
    in_df.ymax = int((y+h/2)  * img.shape[0])
    in_df['class'] = in_df['class'].apply(lambda x: LABELS[x])
    img =cv2.rectangle(img,(in_df.xmin[0],in_df.ymin[0]),(in_df.xmax[0],in_df.ymax[0]),(0,255,255),2)
 #   print(in_df['class'][0])
    cv2.putText(img, 
                in_df['class'][0],
               (in_df.xmin[0], in_df.ymax[0]), 
                font, 1, 
                (0, 255, 255), 
                2, 
                cv2.LINE_4)
    out.write(img)
 #  print(f'Image {img_path}: width={img.shape[1]} height = {img.shape[0]}')
    df = pd.concat([df,in_df])
print(df.head())
df.to_csv(os.path.join(DST,file[:-11]+'_left_bbm.csv'),index = False)
out.release()
```
