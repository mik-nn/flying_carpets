```python
#Библиотеки
import pandas as pd
import re
import os
import shutil
import numpy as np
import cv2
import math
import matplotlib.pyplot as plt
```

```python
#Сервисные функции
def bbox_YOLO(bbox):
    ybox = []
    for i in range(len(bbox)):
        if len(bbox[i]) != 5:
            print('Invalid Dimmension', len(bbox))
            return np.null
        ybox.append([bbox[i][0],(bbox[i][3]+bbox[i][1])//2,(bbox[i][4]+bbox[i][2])//2,bbox[i][3]-bbox[i][1],bbox[i][4]-bbox[i][2]])
    return ybox

%matplotlib inline
def show_images(cols, *images):

    num_images = len(images)
    fig, axes = plt.subplots(cols, math.ceil(num_images/cols), figsize=(10, 5))
    axs = np.ravel(axes)
    for i, image in enumerate(images):
        image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        axs[i].imshow(image, cmap=plt.cm.gray)
        axs[i].axis('off')

    plt.tight_layout()
    plt.show()
    plt.close()
    
def append_instance(img,label):
    out_img = os.path.join(YOLO_DATASET,'Images',img)
    shutil.copy2(img, out_img)
 #   with open()
```

```python
def add_dataset(labelfiles):
    df =pd.read_csv(labelfiles, sep=';')
    df.filename = df.filename.str.replace(r'_\d{6}_jpg.*$','', regex=True)  # Из имен делаем путь к картинкам
    df.head()
    df['class'].value_counts()          #Считаем количество ковров каждого класса
    # Список классов 
    fclasses = df['class'].unique().tolist()
    print(fclasses)
    filenames = df['filename'].unique().tolist()
    images = []
    labels = []
    # Инициализируем видеопо первой строке базы
    i, row = next(df.iterrows())
    filename = str(row['filename'])
    video = os.path.join(VIDEO_PATH,filename,filename+'.mp4')
    cap = cv2.VideoCapture(video)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = cap.get(cv2.CAP_PROP_FPS)
    out_name = os.path.join(VIDEO_PATH,filename,filename+'_bb.mp4')
    out = cv2.VideoWriter(out_name,fourcc, fps, (width,height))
    imgfnum = 1
    if not out:
        print(f'File {out_name} not opened')
        return np.nan
    for index ,row in df.iterrows():
    #for img_f in filenames:
        img_f = row['filename']
        fname = re.sub(r'_\d{6}','',filename)
    #    print(fname)
        fnum = re.findall(r'\d{6}',filename)[0]        # номер кадра в таблице
        for i in range(imgfnum,int(fnum)-1):
            ok, img = cap.read()
            out.write(img)
    #    print(fnum)
    #    for img_p in os.listdir(os.path.join(IMAGE_PATH,img_f)):
    #        imgfnum = re.findall(r'\d{6}',img_p)[0] 
    #        if  imgfnum == fnum:
        bbox =[fclasses.index(row['class']),row['xmin'], row['ymin'],row['xmax'],row['ymax']]
        img_path = os.path.join(IMAGE_PATH,fname,filename+'.jpg')
        images.append(img_path)
        labels.append(bbox)
        print(f'{img_path} class:{fclasses[bbox[0]]},{bbox[1:]}')
        #img = cv2.imread(img_path)
                
        img =cv2.rectangle(img,(bbox[1],bbox[2]),(bbox[3],bbox[4]),(0,255,255),2)
            #    show_images(1,img)
    #       i, row = next(df.iterrows())
        out.write(img)
        
    cap.release()
    out.release()
```
